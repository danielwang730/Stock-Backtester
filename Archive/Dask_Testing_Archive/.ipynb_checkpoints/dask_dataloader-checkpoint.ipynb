```python
# DW using .partitions[n] to get partition number and then computing that
ddf_msft_DW.partitions[13].compute()

# Oh, I think I get it now; basically, the entire ddf dask dataframe is loaded in, and partitions just gets a slice of that whole df
# MSFT happens to be towards the end, and it happens to be split between partitions 13 and 14 (or n = 12 and n = 13)
# So doing ddf_msft.compute() will get you all 1.2 million rows of MSFT, but doing ddf_msft.partitions[13].compute() will only get you the
# last 801k rows for MSFT, and ddf_msft.partitions[12].compute() will get you the 415k rows of MSFT in the 13th partition (n = 12)
```

```python
def contains_tsla(ddf):
    return ddf['ticker'].eq('TSLA').any()

partitions_with_tsla = ddf.map_partitions(contains_tsla).compute()
partitions_with_tsla_indices = [i for i, contains in enumerate(partitions_with_tsla) if contains]
print(partitions_with_tsla_indices)
```

```python
partitions_8_9 = pd.concat([ddf.get_partition(8).compute(), ddf.get_partition(9).compute()])
#print(partitions_8_9)
tsla_data = partitions_8_9.loc[(partitions_8_9['ticker'] == 'TSLA')]
print(tsla_data)
```

```python
ddf_tsla = ddf.loc[ddf['ticker'] == 'TSLA']
ddf_tsla = ddf_tsla.drop(columns=['Unnamed: 0', 'index'])
ddf_tsla = ddf_tsla.rename(columns={'t': 'date'})
ddf_tsla['date'] = dd.to_datetime(ddf_tsla['date'])
ddf_tsla = ddf_tsla.set_index('date')
ddf_tsla.compute()
```
